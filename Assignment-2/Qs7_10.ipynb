***Question 7: Get the IMDb rating for the fifth movie of the dataframe.***

```python
#Answer
ratingno5 = imdb_data.iloc[4]['imdbRating']
print("The IMDb rating for the fifth movie of the dataframe:",ratingno5)
```

***Question 8: Return titles of movies with shortest and longest run time.***

```python
#Answer
shortest = imdb_data.loc[imdb_data['duration'].idxmin()]['title']
longest = imdb_data.loc[imdb_data['duration'].idxmax()]['title']
print("Shortest Movie: ",shortest)
print("Longest Movie: ", longest)
```

***Question 9: Sort the data frame by the order of when they were released and have higher ratings.***

```python
#Answer
sorted = imdb_data.sort_values(by=['year', 'imdbRating'], ascending=[True, False])
sorted.head()
```
